## 1. Creating the Maze Environment

```python
# Grid Configuration
MAZE_SIZE = 15
START = (0, 0)
GOAL = (MAZE_SIZE-1, MAZE_SIZE-1)

# Create World
world = GridWorld.create_maze("default", MAZE_SIZE)

print(f"Grid Created: {MAZE_SIZE}x{MAZE_SIZE}")
```

## 2. Visualizing the Maze (Graph 1)

```python
PathVisualizer.visualize_world(world, start=START, goal=GOAL, title="Lab 8: Default Maze Layout")
```

## 3. Running Experiments (Metrics Calculation)

```python
# Initialize Evaluator
evaluator = HeuristicEvaluator(world, START, GOAL)

# Run Comparison
print("Running Algorithms...")
results = evaluator.compare_all()

# Convert to DataFrame for display
df = pd.DataFrame(results)
df
```

## 4. Performance Comparison Graphs (Graph 2 & 3)

```python
# Plotting Data
labels = [r['heuristic'].replace('heuristic_', '') for r in results]
expansions = [r['expansions_mean'] for r in results]
times = [r['time_mean'] for r in results]

# Create Subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Plot 1: Node Expansions
ax1.bar(labels, expansions, color='skyblue')
ax1.set_title('Node Expansions (Efficiency)', fontweight='bold')
ax1.set_ylabel('Number of Nodes Expanded')
ax1.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels on bars
for i, v in enumerate(expansions):
    ax1.text(i, v + 5, str(int(v)), ha='center', va='bottom', fontweight='bold')

# Plot 2: Execution Time
ax2.bar(labels, times, color='orange')
ax2.set_title('Execution Time (ms)', fontweight='bold')
ax2.set_ylabel('Time in Milliseconds')
ax2.grid(axis='y', linestyle='--', alpha=0.7)

# Add labels on bars
for i, v in enumerate(times):
    ax2.text(i, v + 0.1, f"{v:.1f}ms", ha='center', va='bottom', fontweight='bold')

plt.suptitle('Lab 8: Heuristic Search Performance Comparison', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.show()
```
